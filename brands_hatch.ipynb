```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

# IPython magic to create plots within cells
%matplotlib inline 

# plot style
matplotlib.style.use('ggplot')
```

```python
df = pd.read_csv('data/20171230_163321_bh_win.csv')
```

```python
df2 = df
```

```python

sns.pairplot(x_vars=["track"], y_vars=["speed"], data=df2, \
             hue="lap", aspect=2, size=5, palette='copper')
plt.title('Brands Hatch Speed Analysis [By Lap] (Dec 2017)')
plt.xlim(-50,4000)
```

```python

sns.pairplot(x_vars=["track"], y_vars=["speed"], data=df2, \
             hue="sector", aspect=2, size=5, palette='brg')
plt.title('Brands Hatch Speed Analysis [By Sector] (Dec 2017)')
plt.xlim(-50,4000)
```

```python
sns.boxplot(x='lap', y='speed', data=df2, palette='Set1')
plt.title('Speed by Lap')
```

```python
sns.boxplot(x='lap', y='speed', hue='sector', data=df2, palette='RdYlGn')
plt.title('Speed by Lap/Sector')
```
